# Import thЖ° viб»Үn

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

# Import dб»Ҝ liб»Үu

```python
df = pd.read_csv("temperatures.csv")
df
```

# Chuбә©n hГіa

```python
f_input = df["Farenheit"].to_numpy()
c_output = df["Celcius"].to_numpy()
type(f_input), type(c_output)
```

# Train mГҙ hГ¬nh

```python
# khai bao mo hinh
model = keras.Sequential([ # mo hinh Sequential (tuan tu)
    keras.layers.Dense(units=10, input_shape=[1]), # layer Dense 1 voi 10 neron -> input_shape la 1 vi tra ve gia tri C
    keras.layers.Dense(units=1) # layer Dense 2 voi 1 neron -> tra ve gia tri C
])

model.compile(
    loss = "mean_squared_error", # ham mat mat la mean_squared_error
    optimizer = "adam"
)

model.summary()
```

```python
# train mo hinh
# EPOCH: so lan huan luyen tren du lieu 
model.fit(f_input, c_output, epochs=100)
```

# Test mГҙ hГ¬nh

```python
f_user_input = np.array([50])

# thuc te 
c_output_real = round((f_user_input[0] - 32) / 1.8, 2)
print(f"Gia tri C thuc te: {c_output_real}В°C")

# du doan 
c_output_pred = model.predict(f_user_input)
print(c_output_pred)
print(f"Gia tri C du doan: {c_output_pred[0][0]:.2f}В°C")
```
